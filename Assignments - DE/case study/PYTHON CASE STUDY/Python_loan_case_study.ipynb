```python
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df['LoanAmount'].fillna(df['LoanAmount'].mean(), inplace=True)
df.head()
```

```python
df1 = pd.read_csv("/content/LoanData (1) (1).csv")
df2 = pd.read_csv("/content/LoanData (1) (1).csv")

df = pd.merge(df1, df2)
print(df)
```

```python
df['LoanAmount_k'] = df['LoanAmount'].apply(lambda x: x / 1000)
df.head()
```

```python
df['IncomeLevel'] = df['ApplicantIncome'].apply(lambda x: 'High' if x > 5000 else 'Low')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='Loan_Status', data=df)
plt.title('Loan Approval Count')
plt.show()


```
